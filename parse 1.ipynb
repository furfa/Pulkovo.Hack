```python
import pandas as pd

path = "/home/lyect/Desktop/Приложение №1.xlsx"

data1 = pd.read_excel(path, header=None)
data1.drop([0, 1, 2, 3, 114], axis=1, inplace=True)
data1.drop([3, 4, 11, 12, 13, 14], inplace=True)

data1 = data1.iloc[2:24]

to_drop = []

for col in data1.columns:
    if int(col)%2:
        to_drop.append(col)
        
data1.drop(to_drop, axis=1, inplace=True)

data1 = data1.reset_index(drop=True)

to_drop = []

for idx in data1.index[1:]:
    if int(idx)%2 == 0:
        to_drop.append(idx)
        
data1.drop(to_drop, inplace=True)

new_cols = []

for i, col in enumerate(data1.columns):
    if i == 0:
        new_cols.append("group")
    elif i == 1:
        new_cols.append("course")
    else:
        new_cols.append(data1[col][0].strip().replace("\n", " "))

data1 = data1.rename(columns={k: v for k, v in zip(data1.columns, new_cols)})
data1.drop(0, inplace=True)
data1 = data1.reset_index(drop=True)

data1["group"] = data1["group"].fillna("")
data1 = data1.fillna("")

for col in data1.columns:
    data1[col] = data1[col].apply(lambda x: x.replace("\n", " "))
```
